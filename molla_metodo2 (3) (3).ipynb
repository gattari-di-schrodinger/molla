```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
import numpy as np
```

```python
m=pd.read_csv('masse_cumulative.csv')
T=pd.read_csv('periodi.csv')
x=pd.read_csv('elongazione.csv')
```

T^2 vs m

```python
T_m=T.mean()/5
T_s=T.std()/(5*np.sqrt(10))
m_m=m.mean()
m_s=m.std()
x_m=x.mean()
x_s=x.std()
```

```python
T_qm=pow(T_m,2)
T_qs=2*T_m*T_s
```

```python
plt.errorbar(m_m,T_qm,T_qs,color='black',uplims=True, lolims=True)
plt.xlabel('massa in kg')
plt.ylabel('periodo^2 in s^2')
```

```python
plt.plot(m,pow(T/10,2),'.',color='black')
plt.xlabel('massa in kg')
plt.ylabel('periodo^2 in s^2')
```

```python
plt.errorbar(m_m,x_m,x_s,color='black',uplims=True, lolims=True)
plt.ylabel('spostamento in m')
plt.xlabel('massa in kg')
```

```python
plt.plot(m,x,'.',color='black')
plt.ylabel('spostamento in m')
plt.xlabel('massa in kg')
```
